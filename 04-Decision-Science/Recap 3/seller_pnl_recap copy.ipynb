```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados fict√≠cios
data = pd.DataFrame({
    'GrLivArea': [800, 1000, 1200, 1500, 1700, 1900, 2000, 2300, 2500, 3000],
    'SalePrice': [100000, 120000, 140000, 160000, 180000, 200000, 220000, 250000, 270000, 320000],
    'PredictedPrice': [95000, 115000, 135000, 155000, 175000, 195000, 215000, 240000, 260000, 310000],
    'OtherPrice': [105000, 125000, 145000, 165000, 185000, 205000, 225000, 255000, 275000, 325000]
})

# Reformatando para long format
data_long = pd.melt(
    data,
    id_vars='GrLivArea',
    value_vars=['SalePrice', 'PredictedPrice'],
    var_name='Tipo',
    value_name='Pre√ßo'
)

# Estilo do Seaborn para melhorar o visual
sns.set(style="whitegrid")

# Plotando os gr√°ficos de dispers√£o com marcadores diferentes
sns.scatterplot(data=data_long, x='GrLivArea', y='Pre√ßo', hue='Tipo', palette=['blue', 'orange'], s=100, marker='o', edgecolor='black')

# Plotando a terceira linha (OtherPrice) com uma linha suave (linha de tend√™ncia)
sns.regplot(data=data, x='GrLivArea', y='OtherPrice', label='OtherPrice', scatter=False, color='pink', line_kws={'linewidth': 3, 'linestyle': '--'})

# Adicionando r√≥tulos e t√≠tulo
plt.xlabel("Living Area (m¬≤)")
plt.ylabel("Sale Price ($)")
plt.title("Compara√ß√£o de Pre√ßos (Real, Previsto e Outro)")

# Adicionando a legenda
plt.legend()

# Exibindo o gr√°fico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados fict√≠cios
data = pd.DataFrame({
    'GrLivArea': [800, 1000, 1200, 1500, 1700, 1900, 2000, 2300, 2500, 3000],
    'SalePrice': [100000, 120000, 140000, 160000, 180000, 200000, 220000, 250000, 270000, 320000],
    'PredictedPrice': [95000, 115000, 135000, 155000, 175000, 195000, 215000, 240000, 260000, 310000],
    'OtherPrice': [105000, 125000, 145000, 165000, 185000, 205000, 225000, 255000, 275000, 325000]
})

# Cria√ß√£o de modelo de regress√£o linear para linha de tend√™ncia (OtherPrice)
X = data['GrLivArea'].values.reshape(-1, 1)
y = data['OtherPrice'].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plotando os gr√°ficos de dispers√£o com Matplotlib
plt.scatter(data['GrLivArea'], data['SalePrice'], label='SalePrice', color='blue', edgecolor='black', s=100, marker='o')  # C√≠rculos azuis
plt.scatter(data['GrLivArea'], data['PredictedPrice'], label='PredictedPrice', color='orange', edgecolor='black', s=100, marker='s')  # Quadrados laranjas

# Plotando a linha de tend√™ncia (OtherPrice) com linha de regress√£o
plt.plot(data['GrLivArea'], y_pred, label='OtherPrice', color='pink', linestyle='--', linewidth=3)  # Linha pontilhada rosa

# Adicionando r√≥tulos e t√≠tulo
plt.xlabel("Living Area (m¬≤)")
plt.ylabel("Sale Price ($)")
plt.title("Compara√ß√£o de Pre√ßos (Real, Previsto e Outro)")

# Adicionando a legenda
plt.legend()

# Ativando a grade
plt.grid(True)

# Exibindo o gr√°fico
plt.show()
```
